# Australian Beurau of Stastics Rest-ful interface

The ABS has a little known rest interface. This notebook demonstrates its use in python. Loosely based on both the requests [quickstart](http://docs.python-requests.org/en/latest/user/quickstart/) and the ABS API [documentation](http://stat.abs.gov.au/itt/r.jsp?api).

This notebook will begin by exploring the ABS REST interface and then some simple plotting of the resulting data. 


```python
# <!-- collapse=True -->

import requests
import json
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as pl
import seaborn as sns

# Print version numbers
for module in [requests, json, np, pd, sns, matplotlib]:
    print('{} : {}'.format(module.__name__, module.__version__))

%matplotlib inline

# Do some styling
sns.set_context(rc={"figure.figsize": (10, 10)})
sns.set_style('white')
cmap = pl.cm.gray_r
```

Get list of databases and convert from JSON to python dictionary, and print display the first one.

```python
url = 'http://stat.abs.gov.au/itt/query.jsp'
r1 = requests.get(url, params={'method':'GetDatasetList'})
print(r1.url)
datasetlist = r1.json()
datasetlist['datasets'][0]
```

Extract the ID of the first dataset

```python
dataid = datasetlist['datasets'][0]['id']
print(datasetlist['datasets'][0])
```

Get the concepts available for that ID.

```python
r2 = requests.get(url, params = dict(method='GetDatasetConcepts', datasetid=dataid))
print(r2.url)
concepts = r2.json()
print('keys = {}'.format(concepts.keys()))
print('concepts = {}'.format(concepts['concepts']))
```

So this data set has two axes, there is the amount of rent and also the total familiy income (FINF).

Lets find out the codes for the `MEASURE` and `FINF` concepts.


```python
# <!-- collapse=True -->

r3 = requests.get(url, params=dict(method='GetCodeListValue', datasetid=dataid, 
                                   concept='MEASURE',format='json'))
print(r3.url)
descriptions = r3.json()
#print(  json.dumps(descriptions['codes'], indent=True))
# Lets construct a dictionary for the 'MEASURE' set.

measureNames = {}
count = 0
for element in descriptions['codes']:
    measureNames[element['code']] = (count, element['description'].replace('$','\$').replace(',','').replace('-',' - '))
    count = count + 1
    
print(measureNames)
```

```python
# <!-- collapse=True -->

r3b = requests.get(url, params=dict(method='GetCodeListValue', datasetid=dataid, 
                                   concept='FINF',format='json'))
print(r3b.url)
descriptions = r3b.json()

finfNames = {}
count = 0
for element in descriptions['codes']:
    finfNames[element['code']] = (count, element['description'].replace('$','\$').replace(',','').replace('-',' - '))
    count += 1
print(finfNames)
```

```python
# <!-- collapse=True -->

r3c = requests.get(url, params=dict(method='GetCodeListValue', datasetid=dataid, 
                                   concept='REGION',format='json'))
print(r3c.url)
descriptions = r3c.json()

regionNames = {}
count = 0
for element in descriptions['codes']:
    regionNames[element['code']] = (count, element['description'].replace('$','\$').replace(',','').replace('-',' - '))
    count += 1
print('The first region is {}'.format(regionNames.items()[0]))
```

## Get Generic Data

Its easier if you consider the data as residing an a table (rather than a matrix), each row of the tables has a column for each concept, and then the observations. So when composing the query, we are finding rows of this table that match. Hence the `and` clasue here is to test that the concept is true for that row. Fortunately there are some concepts that are constant throughout the whole table, so if you find those you can extract the whole data set with one query.

*NOTE* That the `and` query must be in the form `CONCEPT`.`CODE`. 

Lets get the stats for Lismore, its region code is 14850

```python
# First construct a list of codes for the measure
r4 = requests.get(url, params={'method':'GetGenericData','datasetid':dataid,
                                  'and':'REGION.14850','orParent':'REGIONTYPE.LGA',
                                   'format':'json'})
print(r4.url)
data = r4.json()
```

It will be helpful to observe the structure of the data we just got. Note that there are 3 observations at different times

```python
print('the keys are {a}'.format(a=data.keys()))
print('An element is \n')
data['series'][0]
```

We could probably and with `REGIONTYPE.LGA2011` if we wanted to get the whole dataset.





Lets write a function that will take a this dictionary and transform it into a pandas data frame in a record style format.

```python
# <!-- collapse=True -->

def abs_to_pandas(abs_data):
    """
    Convert the results from an abs data request into a pandas data-frame.
    
    :param abs_data: Raw data returned from a request to the abs TTL interface with format 'json' specified.
    :returns: A Pandas Data Frame with one row per concept configuration
    """
    data = abs_data['series']
    # Examine the first element to find out what the headings should be:
    # Lets make empty lists which we can fill up as we go.
    headings = {}
    for c in data[0]['concepts']:
        headings[c['name']] = []
    
    # Now examine the observations, construct appropriate headings here too.
    for o in data[0]['observations']:
        headings[o['Time']] = []
            
    for datum in data:
        for c in datum['concepts']:
            headings[c['name']].append(c['Value'])
                         
        for o in datum['observations']:
            headings[o['Time']].append(o['Value'])
                         
    return pd.DataFrame(headings)
    
    
```

```python
df = abs_to_pandas(data)
print(df.head())
df.describe()
```

Lets do some simple visualisation on this data, but lets get rid of any `TOT` that might skew our statistics

```python
df = df.query('FINF != "TOT" and MEASURE != "TOT"')
df = df.sort(['MEASURE', 'FINF'])
```

We want to make a Hexbin, this requires an x and y column in the dataframe. The MEASURE and FINF fields can be sorted, and mapped to x,y coordinates. We  have already built this mapping with the measureNames and finfNames objects.

```python
# <!-- collapse=True -->

# Add convenience columns to the dataframe.
df['x'] = df.MEASURE.apply(lambda x: measureNames[x][0])
df['y'] = df.FINF.apply(lambda x: finfNames[x][0])
df['measureName'] = df.MEASURE.apply(lambda x: measureNames[x][1])
df['finfName'] = df.FINF.apply(lambda x: finfNames[x][1])

values = df[['FINF','MEASURE','2011']]

values = np.array(values.pivot(index='FINF', columns='MEASURE', values='2011').values, dtype='float')
xx,yy = np.meshgrid(df.x, df.y)

fig = pl.hexbin(x=df.x, y=df.y, C=values.ravel(), gridsize=12, cmap=cmap)
pl.colorbar()
pl.xticks(df.x, df.measureName.tolist(), rotation=70)
pl.yticks(df.y, df.finfName.tolist())
pl.ylabel('Family Income per week')
pl.xlabel('Rent per week')
pl.title('Lismore Rent/income breakdown in 2011')
```

The joint distribution of rent vs income for the Lismore region. 
We can also find the marginal distribution as well.

```python
# <!-- collapse=True -->

df['2011'] = df['2011'].astype('float')
pl.figure()
measureMarginal = df[['measureName','2011']].groupby('measureName', sort=False).sum()
print(measureMarginal.shape)
pl.bar(range(len(measureNames)-1), height=measureMarginal.values)
pl.xticks(range(len(measureNames)-1), measureMarginal.index.tolist(), rotation=70)
#pl.yticks(df.y, df.finfName.tolist())

pl.title('Marginal rent per week in Lismore in 2011')

pl.figure()
finfMarginal =  df[['finfName','2011']].groupby('finfName', sort=False).sum()
pl.bar(range(len(finfNames)-1), height=finfMarginal.values)
pl.xticks(range(len(finfNames)-1), finfMarginal.index.tolist(), rotation=70);
pl.title('Marginal family income per week in Lismore in 2011')

```

We could also explore the income change over time.

```python
# <!-- collapse=True -->
# These columns are not numeric, so we must cast them first.
df[['2001', '2006', '2011']] = df[['2001', '2006', '2011']].astype('float')
finfMarginal =  df[['finfName','2001','2006', '2011']].groupby('finfName', sort=False).sum()
finfMarginal.plot(kind='bar')
```
